```python
# Calculating the percentage of outliers values in each variable
outliers_count = {}

for variable in variables_to_plot:
    Q1 = df_tayko_pre[variable].quantile(0.25)
    Q3 = df_tayko_pre[variable].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers_count[variable] = len(df_tayko_pre[(df_tayko_pre[variable] < lower_bound) | (df_tayko_pre[variable] > upper_bound)]) / len(df_tayko_pre[variable])

print(outliers_count)
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training, validation, and test sets
train_data, remaining_data = train_test_split(df_tayko[selected_var_1 + ['Purchase']], test_size=0.40, random_state=2023)
validation_data, test_data = train_test_split(remaining_data, test_size=0.15*len(df_tayko[selected_var_1 + ['Purchase']])/len(remaining_data), random_state=42)

# Separate features (X) and target (y) for each set
X_train_1, y_train_1 = train_data.drop(columns=['Purchase']), train_data['Purchase']
X_validation_1, y_validation_1 = validation_data.drop(columns=['Purchase']), validation_data['Purchase']
X_test_1, y_test_1 = test_data.drop(columns=['Purchase']), test_data['Purchase']
```

```python
import statsmodels.api as sm

# Add a constant to the independent variables (intercept)
X_train_1 = sm.add_constant(X_train_1)
X_validation_1 = sm.add_constant(X_validation_1)
X_test_1 = sm.add_constant(X_test_1)

# Create and fit the logistic regression model
logistic_model_1 = sm.Logit(y_train_1, X_train_1, method='lbfgs')
logistic_result_1 = logistic_model_1.fit()

# Make predictions on the validation set
y_pred_1_pre = logistic_result_1.predict(X_validation_1)

# Choose a threshold for classification (e.g., 0.5)
threshold = 0.5

# Convert probabilities to binary predictions
y_pred_1 = (y_pred_1_pre > threshold).astype(int)

# Get the summary of the logistic regression model
summary_1 = logistic_result_1.summary()

# Print the summary
print(summary_1)
```

### **4. Feature Selection/Engineering** &#x1F680; <a id="item-ten"></a>

##### 4.1. Asociation between variables:

##### *4.1.1. Qualitative Var vs. Qualitative Var:*

```python
from scipy.stats import chi2_contingency

# Assuming df_tayko is your DataFrame
categorical_target = 'Purchase'  # Replace with your target variable name

feature_relation = {}

for var in categorical_var:
    contingency_table = pd.crosstab(df_tayko[var], df_tayko[categorical_target])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    # Set a significance level (alpha) to determine the relationship
    alpha = 0.05  # You can adjust this based on your preference
    
    if p < alpha:
        feature_relation[var] = 1  # Significant relationship (1)
    else:
        feature_relation[var] = 0  # No significant relationship (0)

# Print the result
print("Feature Relation (1: Significant, 0: Not Significant):")
for var, relation in feature_relation.items():
    print(f"{var}: {relation}")
```

##### *4.1.2. Qualitative Var vs. Quantitative Var:*

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Your target variable
target_variable = 'Purchase'

# List of numerical attributes you want to test
numerical_attributes = ['numerical_var1', 'numerical_var2', 'numerical_var3']

# Create an empty list to store ANOVA results
anova_results = []

# Perform ANOVA for each numerical attribute
for attribute in numerical_var:
    formula = f'{attribute} ~ C({target_variable})'
    model = ols(formula, data=df_tayko).fit()
    anova_table = sm.stats.anova_lm(model)
    anova_results.append((attribute, anova_table))

# Print or manipulate the ANOVA results as needed
for attribute, result in anova_results:
    print(f"ANOVA for {attribute}:\n{result}\n")

# Print the p-values for each ANOVA result
for attribute, result in anova_results:
    p_value = result.loc['C(Purchase)', 'PR(>F)']
    print(f"ANOVA for {attribute} - p-value: {p_value:.4f}")
```
